# Bài 3: Thực hành tính sai số

Trong bài này, chúng ta sẽ tìm hiểu về các khái niệm cơ bản liên quan đến phép đo trực tiếp, phép đo gián tiếp, và cách tính toán sai số trong các phép đo. Ngoài ra, chúng ta sẽ sử dụng thư viện Matplotlib để minh họa các kết quả đo lường.

## 1. Phép đo trực tiếp và phép đo gián tiếp

### Phép đo trực tiếp
Phép đo trực tiếp là phép đo mà kết quả được xác định trực tiếp từ dụng cụ đo mà không cần thông qua bất kỳ phép tính nào. Ví dụ:
- Đo chiều dài của một vật bằng thước kẻ.
- Đo nhiệt độ bằng nhiệt kế.

### Phép đo gián tiếp
Phép đo gián tiếp là phép đo mà kết quả được xác định thông qua các phép tính toán dựa trên các phép đo trực tiếp khác. Ví dụ:
- Tính vận tốc của một vật bằng cách đo quãng đường và thời gian, sau đó sử dụng công thức $ v = \frac{s}{t} $.
- Tính diện tích hình chữ nhật bằng cách đo chiều dài và chiều rộng, sau đó sử dụng công thức $ A = l \times w $.

## 2. Sai số phép đo

### Sai số hệ thống
Sai số hệ thống là sai số do các yếu tố cố định gây ra, chẳng hạn như sai số của dụng cụ đo hoặc điều kiện môi trường. Sai số này thường có thể được hiệu chỉnh hoặc loại bỏ.

### Sai số ngẫu nhiên
Sai số ngẫu nhiên là sai số do các yếu tố ngẫu nhiên gây ra, chẳng hạn như sự dao động trong quá trình đo. Sai số này không thể loại bỏ hoàn toàn nhưng có thể giảm thiểu bằng cách lặp lại phép đo nhiều lần.

### Công thức tính sai số
Giả sử chúng ta có một phép đo gián tiếp $ y = f(x_1, x_2, \dots, x_n) $, sai số của $ y $ được tính bằng công thức:

$$
\Delta y = \sqrt{\left(\frac{\partial f}{\partial x_1} \Delta x_1\right)^2 + \left(\frac{\partial f}{\partial x_2} \Delta x_2\right)^2 + \dots + \left(\frac{\partial f}{\partial x_n} \Delta x_n\right)^2}
$$

Trong đó:
- $ \Delta x_i $ là sai số của phép đo trực tiếp $ x_i $.
- $ \frac{\partial f}{\partial x_i} $ là đạo hàm riêng của hàm $ f $ theo biến $ x_i $.


```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu đo lường
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Sai số ngẫu nhiên (luôn dương)
error = np.abs(0.1 * np.random.randn(100))

# Vẽ biểu đồ
plt.figure(figsize=(8, 6))
plt.errorbar(x, y, yerr=error, fmt='o', color='b', ecolor='r', capsize=5, label='Dữ liệu đo với sai số')
plt.plot(x, np.sin(x), 'g--', label='Giá trị lý thuyết')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Biểu đồ minh họa sai số phép đo')
plt.legend()
plt.grid(True)
plt.show()
```
