**Thực Hành Đo Tốc Độ**
=====================

### 1. Cách Đo Tốc Độ Trong Phòng Thí Nghiệm

Để đo tốc độ trong phòng thí nghiệm, chúng ta cần sử dụng các thiết bị đo lường như đồng hồ bấm giờ, thước đo quãng đường và máy tính để xử lý dữ liệu.

**Cách Đo**

* Sử dụng đồng hồ bấm giờ để đo thời gian chuyển động của vật.
* Sử dụng thước đo quãng đường để đo quãng đường di chuyển của vật.
* Sử dụng máy tính để tính toán tốc độ bằng cách chia quãng đường cho thời gian.

### 2. Giới Thiệu Dụng Cụ Đo Thời Gian

Dụng cụ đo thời gian thường được sử dụng trong phòng thí nghiệm là đồng hồ bấm giờ. Đồng hồ bấm giờ có thể được cấu tạo từ các thành phần cơ khí hoặc điện tử, và thường có độ chính xác cao.

**Thiết Bị**

* Đồng hồ bấm giờ: Đây là thiết bị chính để đo thời gian chuyển động của vật.
* Thước đo quãng đường: Sử dụng để đo quãng đường di chuyển của vật.
* Máy tính: Sử dụng để tính toán tốc độ bằng cách chia quãng đường cho thời gian.

### 3. Thực Hành Đo Tốc Độ Chuyển Động

Thực hành này sẽ giúp bạn hiểu rõ hơn về cách đo tốc độ trong phòng thí nghiệm và cách sử dụng các thiết bị đo lường để đạt được kết quả chính xác.

**Thiết Bị**

* Đồng hồ bấm giờ
* Thước đo quãng đường
* Máy tính
* Vật chuyển động (ví dụ: xe nhỏ)

**Cách Thực Hành**

1. Chuẩn bị đồng hồ bấm giờ và thước đo quãng đường.
2. Đặt vật chuyển động trên bề mặt phẳng.
3. Bắt đầu đồng hồ bấm giờ khi vật bắt đầu chuyển động.
4. Sử dụng thước đo quãng đường để đo quãng đường di chuyển của vật.
5. Tính toán tốc độ bằng cách chia quãng đường cho thời gian.

**Bảng Kết Qủa Thí Nghiệm**

| Quãng Đường (m) | Thời Gian (s) | Tốc Độ (m/s) |
| --- | --- | --- |
| 10 | 2 | 5 |
| 20 | 4 | 5 |
| 30 | 6 | 5 |

**Đồ Thị Vẽ Bằng Matplotlib**

Dưới đây là đồ thị vẽ bằng Matplotlib để minh hoạ về mối quan hệ giữa quãng đường và thời gian:


```python
import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng dữ liệu cho quãng đường, thời gian và tốc độ
quang_duong = np.array([10, 20, 30])
thoi_gian = np.array([2, 4, 6])
toc_do = np.array([5, 5, 5])

# Vẽ đồ thị quãng đường - thời gian
plt.plot(quang_duong, thoi_gian, label='Quãng Đường - Thời Gian')
plt.xlabel('Quãng Đường (m)')
plt.ylabel('Thời Gian (s)')
plt.title('Quãng Đường - Thời Gian')
plt.grid(True)
plt.legend()
plt.show()

# Vẽ đồ thị quãng đường - tốc độ
plt.plot(quang_duong, toc_do, label='Quãng Đường - Tốc Độ')
plt.xlabel('Quãng Đường (m)')
plt.ylabel('Tốc Độ (m/s)')
plt.title('Quãng Đường - Tốc Độ')
plt.grid(True)
plt.legend()
plt.show()

# Vẽ đồ thị thời gian - tốc độ
plt.plot(thoi_gian, toc_do, label='Thời Gian - Tốc Độ')
plt.xlabel('Thời Gian (s)')
plt.ylabel('Tốc Độ (m/s)')
plt.title('Thời Gian - Tốc Độ')
plt.grid(True)
plt.legend()
plt.show()
```
