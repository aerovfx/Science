

**Sự rơi tự do**

**1. Sự rơi trong không khí**

Khi một vật được thả từ trên cao xuống, nó sẽ bắt đầu rơi xuống dưới do tác dụng của trọng lực. Sự rơi này gọi là sự rơi trong không khí.

Trong thực tế, không khí tạo ra một số trở ngại cho sự rơi tự do của vật, chẳng hạn như:

* Áp suất không khí: Áp suất không khí sẽ tạo ra một lực ngược chiều hướng xuống của vật.
* Sức cản của không khí: Sức cản của không khí sẽ làm giảm vận tốc của vật.

Mặc dù có những trở ngại này, sự rơi trong không khí vẫn là một quá trình phức tạp và cần được nghiên cứu kỹ lưỡng.

**2. Sự rơi tự do**

Sự rơi tự do là một loại chuyển động mà vật chỉ bị tác dụng bởi trọng lực, không có bất kỳ lực nào khác ngoài trọng lực. Trong trường hợp này, vật sẽ rơi xuống dưới với gia tốc 9,8 m/s^2.

Đồ thị vận tốc - thời gian của sự rơi tự do:

| Thời gian (s) | Vận tốc (m/s) |
| --- | --- |
| 0 | 0 |
| 1 | 9,8 |
| 2 | 19,6 |
| 3 | 29,4 |
| 4 | 39,2 |

Và dưới đây là mã Python để vẽ đồ thị vận tốc - thời gian bằng Matplotlib:



````python

import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng dữ liệu cho vận tốc và thời gian
van_toc = np.array([0, 9.8, 19.6, 29.4, 39.2])
thoi_gian = np.array([0, 1, 2, 3, 4])

# Vẽ đồ thị vận tốc - thời gian
plt.plot(thoi_gian, van_toc)
plt.xlabel('Thời Gian (s)')
plt.ylabel('Vận Tốc (m/s)')
plt.title('Đồ Thị Vận Tốc - Thời Gian')
plt.grid(True)
plt.show()
````

```

Kết quả sẽ là một đồ thị thẳng đứng, thể hiện sự tăng dần của vận tốc trong thời gian.
