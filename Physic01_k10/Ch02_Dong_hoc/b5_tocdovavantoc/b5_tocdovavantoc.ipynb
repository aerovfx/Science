**Tốc Độ Và Vận Tốc**
=====================

### 1. Tốc Độ

Tốc độ là đại lượng đo lường sự thay đổi của vị trí của một vật trong một đơn vị thời gian.

**Định Nghĩa**

* Là đại lượng đo lường sự thay đổi của vị trí của một vật trong một đơn vị thời gian.
* Được biểu diễn bằng đơn vị đo dài, như mét hoặc kilômét, và đơn vị đo thời gian, như giây hoặc phút.

**Ví Dụ**

Khi một xe ô tô di chuyển với tốc độ 60 km/h, điều đó có nghĩa là xe sẽ đi được quãng đường 1 km trong khoảng thời gian 16,7 s (do 1 giờ = 3600 s và 60 km/h = 60000 m/3600 s).

### 2. Vận Tốc

Vận tốc là đại lượng đo lường sự thay đổi của vị trí của một vật trong một đơn vị thời gian.

**Định Nghĩa**

* Là đại lượng đo lường sự thay đổi của vị trí của một vật trong một đơn vị thời gian.
* Được biểu diễn bằng vectơ, với hướng và độ lớn.

**Ví Dụ**

Khi một xe ô tô di chuyển với vận tốc 60 km/h về phía Đông, điều đó có nghĩa là xe sẽ đi được quãng đường 1 km về phía Đông trong khoảng thời gian 16,7 s.

### Quãng Đường - Thời Gian

Quãng đường và thời gian là hai đại lượng liên quan đến nhau qua tốc độ. Khi biết quãng đường và thời gian, ta có thể tính được tốc độ của vật.

**Ví Dụ**

Khi một xe ô tô di chuyển với quãng đường 1 km và thời gian 16,7 s, ta có thể tính được tốc độ của xe như sau:

Tốc độ = Quãng đường / Thời gian
= 1000 m / 16,7 s
= 60 km/h

### Độ Dịch Chuyển - Thời Gian

Độ dịch chuyển và thời gian là hai đại lượng liên quan đến nhau qua vận tốc. Khi biết độ dịch chuyển và thời gian, ta có thể tính được vận tốc của vật.

**Ví Dụ**

Khi một xe ô tô di chuyển với độ dịch chuyển 1000 m và thời gian 16,7 s, ta có thể tính được vận tốc của xe như sau:

Vận tốc = Độ dịch chuyển / Thời gian
= 1000 m / 16,7 s
= 60 km/h



**Bảng Đồ Thư Vẽ Bằng Matplotlib**

Dưới đây là bảng đồ thị vẽ bằng Matplotlib để minh hoạ về quãng đường - thời gian và độ dịch chuyển - thời gian:


```python

import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng dữ liệu cho quãng đường và thời gian
quang_duong = np.array([0, 100, 200, 300, 400])
thoi_gian = np.array([0, 10, 20, 30, 40])

# Vẽ đồ thị quãng đường - thời gian
plt.plot(quang_duong, thoi_gian)
plt.xlabel('Quãng Đường (m)')
plt.ylabel('Thời Gian (s)')
plt.title('Quãng Đường - Thời Gian')
plt.show()

# Tạo mảng dữ liệu cho độ dịch chuyển và thời gian
do_dich_chuyen = np.array([0, 100, 200, 300, 400])
thoi_gian = np.array([0, 10, 20, 30, 40])

# Vẽ đồ thị độ dịch chuyển - thời gian
plt.plot(do_dich_chuyen, thoi_gian)
plt.xlabel('Độ Dịch Chuyển (m)')
plt.ylabel('Thời Gian (s)')
plt.title('Độ Dịch Chuyền - Thời Gian')
plt.show()
```
