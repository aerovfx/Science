**Đồ thị độ dịch chuyển - thời gian**

Khi một vật chuyển động, chúng ta có thể sử dụng đồ thị độ dịch chuyển - thời gian để mô tả sự thay đổi của vị trí vật trong thời gian. Đồ thị này được sử dụng rộng rãi trong vật lý học và là một công cụ quan trọng để phân tích và giải thích các hiện tượng chuyển động.

**1. Chuyển động thẳng**

Chuyển động thẳng là loại chuyển động mà vật di chuyển theo một đường thẳng đều, không có sự thay đổi về hướng hoặc tốc độ. Ví dụ: một xe ô tô di chuyển trên một con đường thẳng, một người chạy trên một đường đua thẳng.

**2. Đồ thị độ dịch chuyển - thời gian trong chuyển động thẳng**

Để mô tả chuyển động thẳng bằng đồ thị độ dịch chuyển - thời gian, chúng ta cần có hai mảng dữ liệu: độ dịch chuyển (x) và thời gian (t). Mảng dữ liệu này thường được biểu diễn trên một hệ tọa độ Cartesian với trục x cho độ dịch chuyển và trục y cho thời gian.

Dưới đây là bảng số liệu ví dụ:

| Thời gian (s) | Độ dịch chuyển (m) |
| --- | --- |
| 0 | 0 |
| 1 | 10 |
| 2 | 20 |
| 3 | 30 |
| 4 | 40 |

Và dưới đây là đồ thị độ dịch chuyển - thời gian tương ứng:


```python

import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng dữ liệu cho độ dịch chuyển và thời gian
do_dich_chuyen = np.array([0, 10, 20, 30, 40])
thoi_gian = np.array([0, 1, 2, 3, 4])

# Vẽ đồ thị độ dich chuyển - thời gian
plt.plot(do_dich_chuyen, thoi_gian)
plt.xlabel('Độ Dịch Chuyển (m)')
plt.ylabel('Thời Gian (s)')
plt.title('Đồ Thị Độ Dịch Chuyển - Thời Gian')
plt.grid(True)
plt.show()
```


**3. Vận tốc và đồ thị độ dịch chuyển - thời gian trong chuyển động thẳng**

Vận tốc là một đặc điểm quan trọng của chuyển động, mô tả sự thay đổi về vị trí của vật trong một đơn vị thời gian. Trong chuyển động thẳng, vận tốc thường giữ nguyên.

Dưới đây là bảng số liệu ví dụ:

| Thời gian (s) | Độ dịch chuyển (m) | Vận tốc (m/s) |
| --- | --- | --- |
| 0 | 0 | 0 |
| 1 | 10 | 10 |
| 2 | 20 | 10 |
| 3 | 30 | 10 |
| 4 | 40 | 10 |

Và dưới đây là đồ thị độ dịch chuyển - thời gian và vận tốc tương ứng:


```python

import matplotlib.pyplot as plt
import numpy as np

# Tạo mảng dữ liệu cho độ dich chuyển, thời gian và vận tốc
do_dich_chuyen = np.array([0, 10, 20, 30, 40])
thoi_gian = np.array([0, 1, 2, 3, 4])
van_toc = np.array([0, 10, 10, 10, 10])

# Vẽ đồ thị độ dich chuyển - thời gian
plt.figure(figsize=(10,5))

plt.subplot(121)
plt.plot(do_dich_chuyen, thoi_gian)
plt.xlabel('Độ Dịch Chuyển (m)')
plt.ylabel('Thời Gian (s)')
plt.title('Đồ Thị Độ Dịch Chuyền - Thời Gian')

# Vẽ đồ thị vận tốc
plt.subplot(122)
plt.plot(van_toc, thoi_gian)
plt.xlabel('Vận Tốc (m/s)')
plt.ylabel('Thời Gian (s)')
plt.title('Đồ Thị Vận Tốc - Thời Gian')

plt.tight_layout()
plt.grid(True)
plt.show()

```
