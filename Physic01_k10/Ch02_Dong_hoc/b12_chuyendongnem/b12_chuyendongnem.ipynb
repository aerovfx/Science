**1. Chuyển động ném ngang**

Chuyển động ném ngang là một loại chuyển động mà vật được ném ra khỏi điểm xuất phát với góc độ ngang và không có thành phần hướng lên trên hoặc xuống dưới.

Các yếu tố ảnh hưởng đến chuyển động ném ngang:

* Vận tốc ban đầu của vật
* Góc ném (góc giữa đường ném và mặt đất)
* Hướng ném (phải trái, phải, trước hay sau)




### Ném ngang

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo mảng dữ liệu cho thời gian từ 0 đến 4 giây với bước 0,1 s
t = np.arange(0, 4.1, 0.1)

# Tính toán độ cao (H) và khoảng cách (L)
g = -9.8  # gia tốc trọng trường
v_0 = 10  # vận tốc ban đầu

H = 0.5 * g * t**2
L = v_0 * np.sqrt(2 * H / g)

# Vẽ đồ thị chuyển động ném ngang
plt.plot(L, H)
plt.xlabel('Khoảng cách (m)')
plt.ylabel('Độ cao (m)')
plt.title('Đồ thị chuyển động ném ngang')
plt.grid(True)
plt.show()
```

### Ném xiên

Công thức tầm xa (L) và độ cao (H) trong ném xiên được tính toán như sau:

**Tầm xa (L)**:

L = v0² \* sin(2α) / g

**Độ cao (H)**:

H = v0² \* sin²(α) / (2g)

Trong đó:

* v0 là vận tốc ban đầu
* α là góc ném
* g là gia tốc trọng trường (khoảng 9,8 m/s²)
* L là tầm xa
* H là độ cao

Lưu ý rằng công thức trên chỉ áp dụng cho ném xiên với góc ném α nằm trong khoảng từ 0 đến 90 độ.

## Đồ thị với góc ném alpha

```python
import numpy as np
import matplotlib.pyplot as plt

# Tạo mảng dữ liệu cho thời gian từ 0 đến 4 giây với bước 0,1 s
t = np.arange(0, 4.2, 0.1)

# Tính toán độ cao (H) và khoảng cách (L)
g = -9.8 # gia tốc trọng trường
v_0x = 10 # vận tốc ban đầu theo
v_0y = 20 # vận tốc ban đầu theo y

H = v_0y * t + 0.5 * g * t**2
L = v_0x * t

# Vẽ đồ thị chuyển động ném xiên
plt.plot(L, H)
plt.xlabel('Khoảng cách (m)')
plt.ylabel('Độ cao (m)')
plt.title('Đồ thị chuyển động ném xiên')
plt.grid(True)
plt.show()
```

## Đồ thị để tìm góc ném tối ưu

```python
import numpy as np
import matplotlib.pyplot as plt

# Định nghĩa tham số
v0 = 100  # Vận tốc ban đầu (m/s)
g = 9.8  # Gia tốc trọng trường (m/s²)

# Tạo mảng dữ liệu cho góc ném từ 0 đến 90 độ với bước 1 độ
alpha = np.arange(0, 91, 1)  # Mảng dữ liệu cho góc ném

# Tính toán tầm xa và độ cao
L = v0**2 * np.sin(2*alpha) / g
H = v0**2 * np.sin(alpha)**2 / (2*g)

# Vẽ đồ thị
plt.plot(alpha, L, label='Tầm xa')
plt.plot(alpha, H, label='Độ cao')
plt.xlabel('Góc ném (độ)')
plt.ylabel('Khoảng cách (m)')
plt.title('Đồ thị ném xiên')
plt.grid(True)
plt.legend()
plt.show()
```
