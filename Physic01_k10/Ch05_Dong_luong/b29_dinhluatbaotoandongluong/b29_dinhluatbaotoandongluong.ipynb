# Bài 29: Định Luật Bảo Toàn Động Lượng

## 1. Định Luật Bảo Toàn Động Lượng

:::{note}
**Định luật bảo toàn động lượng** phát biểu rằng nếu tổng ngoại lực tác dụng lên một hệ bằng không, thì tổng động lượng của hệ đó được bảo toàn.
:::

**Biểu thức toán học**:

$$
\mathbf{p}_{\text{trước}} = \mathbf{p}_{\text{sau}}
$$

Tức là:

$$
m_1 \mathbf{v}_1 + m_2 \mathbf{v}_2 = m_1 \mathbf{v}_1' + m_2 \mathbf{v}_2'
$$

Trong đó:
- $m_1, m_2$ là khối lượng của hai vật.
- $\mathbf{v}_1, \mathbf{v}_2$ là vận tốc của hai vật trước va chạm.
- $\mathbf{v}_1', \mathbf{v}_2'$ là vận tốc của hai vật sau va chạm.

**Điều kiện áp dụng**:
- Hệ phải là **hệ kín**, tức là không có ngoại lực tác dụng hoặc tổng ngoại lực bằng không.
- Chỉ xét các **tương tác nội lực** trong hệ.

---

## 2. Va Chạm Đàn Hồi và Va Chạm Mềm

### **2.1. Va Chạm Đàn Hồi**
:::{admonition} Định nghĩa 📌
**Va chạm đàn hồi** là va chạm mà cả **động lượng** và **động năng** đều được bảo toàn.
:::

**Điều kiện**:
- Không có mất mát năng lượng do ma sát, nhiệt độ hoặc biến dạng.
- Tổng động năng trước và sau va chạm không đổi:

$$
\frac{1}{2} m_1 v_1^2 + \frac{1}{2} m_2 v_2^2 = \frac{1}{2} m_1 v_1'^2 + \frac{1}{2} m_2 v_2'^2
$$

🔹 **Ví dụ**: Các nguyên tử và phân tử khí va chạm với nhau trong một hệ cô lập.

---

### **2.2. Va Chạm Mềm**
:::{admonition} Định nghĩa 📌
**Va chạm mềm** là va chạm trong đó hai vật dính vào nhau sau va chạm, nghĩa là chúng có cùng vận tốc sau va chạm.
:::

Trong trường hợp này, chỉ có động lượng được bảo toàn, nhưng **động năng không được bảo toàn** do một phần năng lượng bị chuyển thành nhiệt hoặc làm biến dạng vật.

**Biểu thức động lượng bảo toàn**:

$$
m_1 v_1 + m_2 v_2 = (m_1 + m_2) v'
$$

Trong đó:
- $v'$ là vận tốc chung của hai vật sau va chạm.

🔹 **Ví dụ**: Một viên đạn găm vào một khối gỗ và cả hai cùng chuyển động sau va chạm.

---

```{admonition} Ứng dụng 🎯
- **Pháo phản lực**: Khi một vật bắn ra theo một hướng, phần còn lại của hệ di chuyển theo hướng ngược lại.
- **Tên lửa**: Hoạt động dựa trên nguyên tắc bảo toàn động lượng.
- **Bóng bi-a**: Các va chạm giữa các viên bi tuân theo quy tắc của động lượng bảo toàn.
