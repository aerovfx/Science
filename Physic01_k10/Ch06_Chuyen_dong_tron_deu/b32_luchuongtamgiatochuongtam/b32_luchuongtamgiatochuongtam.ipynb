**Lực Hướng Tâm và Gia Tốc Hướng Tâm**
=====================================

### 1. Lực Hướng Tâm

Lực hướng tâm là một loại lực mà tác dụng lên một vật thể di chuyển trên một đường tròn, khiến cho vật thể luôn giữ được hướng chuyển động của mình.

Lực hướng tâm có đặc điểm sau:

* Được biểu diễn bởi một vectơ
* Có phương hướng tâm
* Có mô men đối diện với hướng di chuyển của vật thể

### 2. Gia Tốc Hướng Tâm

Gia tốc hướng tâm là gia tốc của một vật thể di chuyển trên một đường tròn, được đo bằng đơn vị m/s^2.

Gia tốc hướng tâm có đặc điểm sau:

* Được biểu diễn bởi một vectơ
* Có phương hướng tâm
* Có mô men đối diện với hướng di chuyển của vật thể

### 3. Công Thức Độ Lớn Của Lực Hướng Tâm

Công thức độ lớn của lực hướng tâm được tính bằng công thức sau:

F = (m \* v^2) / r

trong đó:
- F là lực hướng tâm
- m là khối lượng của vật thể
- v là tốc độ của vật thể
- r là bán kính của đường tròn

Lực hướng tâm có đặc điểm sau:

* Tăng lên khi tốc độ của vật thể tăng lên
* Tăng lên khi bán kính của đường tròn giảm xuống
* Giữ nguyên nếu cả tốc độ và bán kính không thay đổi.

**Ví dụ Minh Họa**

Một ô tô di chuyển trên một đường tròn với bán kính 20 m. Tốc độ của ô tô là 50 km/h. Lực hướng tâm có giá trị là?

F = (m \* v^2) / r
= (1000 kg \* (50 km/h)^2) / 20 m
= 125000 N

**Kết Luận**

Trên đây là một số khái niệm chính về lực hướng tâm và gia tốc hướng tâm. Hy vọng rằng người đọc đã hiểu và nhớ được các yếu tố quan trọng của bài học này.

**Lời Gợi Ý cho Bài Tập**

* Tính lực hướng tâm của một vật thể di chuyển trên một đường tròn với bán kính 10 m và tốc độ 30 m/s.
* Tính gia tốc hướng tâm của một vật thể di chuyển trên một đường tròn với bán kính 15 m và tốc độ 40 m/s.
