**Chuyển Động Tròn Đều**
=====================

### Mô tả Chuyển Động Tròn

Chuyển động tròn là loại chuyển động trong đó một vật thể di chuyển trên một đường tròn với tâm cố định. Chuyển động này được đặc trưng bởi việc di chuyển của vật thể trên một khoảng cách xác định quanh một điểm trung tâm.

### Chuyển Động Tròn Đều - Tốc Độ và Tốc Độ Góc

Chuyển động tròn đều là loại chuyển động tròn trong đó tốc độ của vật thể không đổi. Tốc độ của vật thể được đo bằng đơn vị m/s, còn tốc độ góc được đo bằng đơn vị rad/s.

Tốc độ góc (ω) là tốc độ với đó vật thể di chuyển trên đường tròn và được tính bằng công thức:

```latex
ω = Δθ / Δt
```

trong đó:
- ω là tốc độ góc
- Δθ là góc quay của vật thể
- Δt là thời gian cần thiết để vật thể quay qua góc đó

Tốc độ (v) của vật thể trong chuyển động tròn đều được tính bằng công thức:

```latex
v = rω
```

trong đó:
- v là tốc độ
- r là bán kính đường tròn

### Vận Tốc Trong Chuyển Động Tròn Đều

Vận tốc của vật thể trong chuyển động tròn đều là một vectơ có phương hướng tâm và mô men của nó luôn đối diện với hướng di chuyển của vật thể.

Vận tốc được tính bằng công thức:

```latex
v = rω (hướng tâm)
```

hoặc:

```latex
v = -rω (hướng ra ngoài)
```

trong đó:
- v là vận tốc
- r là bán kính đường tròn
- ω là tốc độ góc

**Ví dụ Minh Họa**

Một ô tô di chuyển trên một đường tròn với bán kính 50 m. Tốc độ của ô tô là 30 m/s. Tính tốc độ góc (ω) và vận tốc (v) của ô tô?

Tốc độ góc (ω):
```latex
ω = v / r = 30 m/s / 50 m = 0,6 rad/s
```

Vận tốc (v):
```latex
v = rω = 50 m x 0,6 rad/s = 30 m/s
```
