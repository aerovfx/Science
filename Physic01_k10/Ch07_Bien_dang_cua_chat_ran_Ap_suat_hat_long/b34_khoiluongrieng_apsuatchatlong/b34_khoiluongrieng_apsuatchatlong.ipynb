**Khối Lượng Riêng Và Áp Suất Chất Lỏng**
=====================================

### 1. Khối Lượng Riêng

Khối lượng riêng là đại lượng đặc trưng cho khối lượng của một đơn vị thể tích của một chất lỏng. Đơn vị đo khối lượng riêng thường sử dụng là kg/m^3.

**Tính Toán Khối Lượng Riêng**

Khối lượng riêng (ρ) được tính toán bằng công thức sau:

ρ = m / V

trong đó:
- ρ là khối lượng riêng
- m là khối lượng của chất lỏng
- V là thể tích của chất lỏng

**Ví dụ Minh Họa**

Khối lượng riêng của nước ở 20°C và 1 atm là khoảng 1000 kg/m^3.

### 2. Áp Lực Và Áp Suất

Áp lực (P) là đại lượng đặc trưng cho lực tác dụng lên một đơn vị diện tích của bề mặt chất lỏng. Đơn vị đo áp lực thường sử dụng là Pa (Pascal).

**Tính Toán Áp Suất**

Áp suất (p) được tính toán bằng công thức sau:

p = P / A

trong đó:
- p là áp suất
- P là áp lực
- A là diện tích của bề mặt chất lỏng

**Ví dụ Minh Họa**

Áp suất của nước ở 20°C và 1 atm là khoảng 101325 Pa.

### Đồ Thị Mô tả Sự Thay Đổi Của Khối Lượng Riêng Theo Nhiệt Độ Và Áp Suất Theo Độ Cao (Độ Sâu)

Dưới đây là mã Python sử dụng thư viện matplotlib để tạo đồ thị mô tả sự thay đổi của khối lượng riêng theo nhiệt độ và áp suất theo độ cao (độ sâu):


```python

import numpy as np
import matplotlib.pyplot as plt

# Tạo mảng dữ liệu cho nhiệt độ, áp suất và độ cao
T = np.linspace(0, 100, 100)
P = np.linspace(0, 10, 100)
H = np.linspace(0, 1000, 100)

# Tạo mảng dữ liệu cho khối lượng riêng
rho_T = 1000 + (T / 100) * 5
rho_P = 1000 + (P / 10) * 1
rho_H = 1000 - (H / 1000) * 2

# Tạo đồ thị
plt.figure(figsize=(12,6))

plt.subplot(131)
plt.plot(T, rho_T)
plt.xlabel('Nhiệt độ (°C)')
plt.ylabel('Khối lượng riêng (kg/m^3)')
plt.title('Sự thay đổi của khối lượng riêng theo nhiệt độ')

plt.subplot(132)
plt.plot(P, rho_P)
plt.xlabel('Áp suất (atm)')
plt.ylabel('Khối lượng riêng (kg/m^3)')
plt.title('Sự thay đổi của khối lượng riêng theo áp suất')

plt.subplot(133)
plt.plot(H, rho_H)
plt.xlabel('Độ cao (m)')
plt.ylabel('Khối lượng riêng (kg/m^3)')
plt.title('Sự thay đổi của khối lượng riêng theo độ cao')

plt.tight_layout()
plt.show()
```
