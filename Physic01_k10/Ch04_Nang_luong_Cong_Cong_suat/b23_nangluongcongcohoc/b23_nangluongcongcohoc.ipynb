# Bài 23: Năng Lượng. Công Cơ Học

## 1. Năng Lượng
:::{note}
**Năng lượng** là đại lượng đặc trưng cho khả năng thực hiện công của một vật hoặc hệ vật.
:::

- Một vật có năng lượng khi nó có khả năng thực hiện công hoặc truyền nhiệt.
- Năng lượng tồn tại dưới nhiều dạng:
  - **Cơ năng**: Gồm động năng và thế năng.
  - **Nhiệt năng**: Liên quan đến chuyển động nhiệt của các phân tử.
  - **Điện năng**: Do các hạt mang điện tạo ra.
  - **Hóa năng**: Dự trữ trong liên kết hóa học.
  - **Hạt nhân năng**: Do sự phân hạch hoặc tổng hợp hạt nhân.

---

## 2. Công Cơ Học
:::{note}
**Công cơ học** là công do một lực tác dụng lên vật và làm vật di chuyển theo hướng của lực.
:::

Công cơ học được xác định bởi công thức:

$$
A = F s \cos\theta
$$

Trong đó:
- $ A $ là công thực hiện (J – Joule).
- $ F $ là độ lớn của lực tác dụng (N – Newton).
- $ s $ là quãng đường vật dịch chuyển (m – mét).
- $ \theta $ là góc giữa lực $ F $ và hướng dịch chuyển.

**Các trường hợp đặc biệt:**
1. Nếu $ \theta = 0^\circ $ (lực cùng phương với chuyển động):
   $$
   A = F s
   $$
2. Nếu $ \theta = 90^\circ $ (lực vuông góc với chuyển động):
   $$
   A = 0
   $$
   → Lực không sinh công (ví dụ: lực pháp tuyến của mặt phẳng lên vật).
3. Nếu $ \theta = 180^\circ $ (lực ngược hướng chuyển động):
   $$
   A = - F s
   $$
   → Công âm (lực cản làm giảm động năng của vật).
