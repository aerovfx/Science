# Bài 25: Động Năng Và Thế Năng

## 1. Động Năng

:::{note}
**Động năng** là năng lượng mà một vật có được do chuyển động.
:::

- Mọi vật chuyển động đều có động năng.
- Động năng phụ thuộc vào **khối lượng** và **vận tốc** của vật.
- Công thức tính động năng:

$$
W_d = \frac{1}{2} m v^2
$$

Trong đó:
- $ W_d $ là động năng (J – Joule).
- $ m $ là khối lượng của vật (kg).
- $ v $ là vận tốc của vật (m/s).

**Nhận xét:**
- Khi $ v = 0 $, động năng bằng **0** → vật đứng yên không có động năng.
- Khi $ v $ tăng, động năng tăng.
- Động năng chỉ có giá trị **dương** hoặc **bằng 0**, không có giá trị âm.

---

## 2. Thế Năng

### 2.1 Thế Năng Trọng Trường

:::{tip}
**Thế năng trọng trường** là năng lượng của vật có được do vị trí của nó trong trọng trường.
:::

- Một vật ở **độ cao** càng lớn thì thế năng trọng trường càng lớn.
- Công thức tính thế năng trọng trường:

$$
W_t = m g h
$$

Trong đó:
- $ W_t $ là thế năng trọng trường (J – Joule).
- $ m $ là khối lượng của vật (kg).
- $ g $ là gia tốc trọng trường ($ 9.8 \, m/s^2 $).
- $ h $ là độ cao của vật so với mốc thế năng (m).

**Nhận xét:**
- Khi vật rơi xuống, thế năng giảm, động năng tăng.
- Khi vật được nâng lên, thế năng tăng.

### 2.2 Thế Năng Đàn Hồi

:::{tip}
**Thế năng đàn hồi** là năng lượng tích trữ trong vật đàn hồi khi nó bị biến dạng.
:::

- Ví dụ: **Lò xo bị nén hoặc kéo dãn** có thế năng đàn hồi.
- Công thức tính thế năng đàn hồi:

$$
W_{\text{đh}} = \frac{1}{2} k x^2
$$

Trong đó:
- $ W_{\text{đh}} $ là thế năng đàn hồi (J – Joule).
- $ k $ là độ cứng của lò xo (N/m).
- $ x $ là độ biến dạng của lò xo (m).

---

```{admonition} Lưu ý 📌
- Động năng và thế năng có thể chuyển hóa qua lại nhưng tổng cơ năng luôn được bảo toàn trong hệ kín.
- Khi vật rơi tự do: **Thế năng chuyển hóa thành động năng**.
- Khi ném vật lên cao: **Động năng chuyển hóa thành thế năng**.
