# Bài 26: Cơ Năng và Định Luật Bảo Toàn Cơ Năng

## 1. Sự Chuyển Hoá Giữa Động Năng và Thế Năng

:::{note}
Trong quá trình chuyển động, động năng và thế năng có thể **chuyển hóa lẫn nhau** nhưng tổng cơ năng luôn được bảo toàn trong hệ kín.
:::

- Khi vật **rơi tự do**:
  - Thế năng giảm dần, động năng tăng dần.
  - Tổng cơ năng không đổi.

- Khi vật **ném lên cao**:
  - Động năng giảm, thế năng tăng.
  - Tổng cơ năng vẫn bảo toàn.

- Khi vật dao động như **con lắc**:
  - Tại vị trí cao nhất: Thế năng cực đại, động năng bằng 0.
  - Tại vị trí thấp nhất: Động năng cực đại, thế năng bằng 0.

- Công thức tổng quát:

$$
W = W_d + W_t
$$

Trong đó:
- $ W $ là **cơ năng** của vật.
- $ W_d = \frac{1}{2} m v^2 $ là **động năng**.
- $ W_t = mgh $ là **thế năng trọng trường**.

---

## 2. Định Luật Bảo Toàn Cơ Năng

:::{admonition} Định luật bảo toàn cơ năng 📜
Trong một hệ kín, nếu chỉ có lực thế (trọng lực, lực đàn hồi) tác dụng, thì **tổng cơ năng của vật luôn được bảo toàn**.
:::

$$
W_1 = W_2
$$

Tức là:

$$
W_d1 + W_t1 = W_d2 + W_t2
$$

**Điều kiện áp dụng**:
- Không có lực cản như ma sát, lực cản không khí.
- Chỉ có các lực bảo toàn (trọng lực, lực đàn hồi).

**Ứng dụng của định luật bảo toàn cơ năng**:
1. **Chuyển động rơi tự do**: Khi vật rơi, thế năng chuyển thành động năng.
2. **Dao động con lắc**: Cơ năng bảo toàn trong một chu kỳ dao động.
3. **Chuyển động của viên bi trên máng cong**: Tại mỗi điểm trên quỹ đạo, tổng động năng và thế năng của viên bi không đổi.

---

```{admonition} Lưu ý 🚀
- Nếu có ma sát hoặc lực cản, cơ năng không còn bảo toàn mà sẽ chuyển hóa một phần thành nhiệt năng.
- Định luật bảo toàn cơ năng giúp giải thích nhiều hiện tượng vật lý và ứng dụng trong kỹ thuật.
