# **NĂNG LƯỢNG VÀ CÔNG SUẤT ĐIỆN**

## **1. Năng lượng điện**
- **Năng lượng điện** là năng lượng do dòng điện sinh ra khi nó thực hiện công lên các thiết bị điện.
- Khi dòng điện chạy qua một thiết bị điện có điện trở $ R $, năng lượng tiêu thụ bởi thiết bị chính là công của lực điện thực hiện để dịch chuyển các hạt mang điện.
- **Công của dòng điện** trong khoảng thời gian $ t $ được tính theo công thức:

  $$
  A = UIt
  $$

  hoặc

  $$
  A = R I^2 t
  $$

  Trong đó:
  - $ A $ là công của dòng điện (J).
  - $ U $ là hiệu điện thế giữa hai đầu thiết bị (V).
  - $ I $ là cường độ dòng điện chạy qua thiết bị (A).
  - $ t $ là thời gian dòng điện chạy qua thiết bị (s).
  - $ R $ là điện trở của thiết bị (Ω).

- **Đơn vị của năng lượng điện**:
  - **Joule (J)**: Đơn vị chuẩn trong hệ SI.
  - **Wh (Watt-hour)**: 1 Wh = 3600 J.
  - **kWh (Kilowatt-hour)**: 1 kWh = 1000 Wh = $ 3.6 \times 10^6 $ J.

- **Ứng dụng**:
  - Dùng để tính toán mức tiêu thụ điện năng trong các thiết bị gia dụng như bóng đèn, tivi, máy lạnh, v.v.
  - Công tơ điện đo năng lượng điện tiêu thụ theo đơn vị **kWh**.

---

## **2. Công suất điện**
- **Công suất điện** là đại lượng đo tốc độ tiêu thụ năng lượng điện của một thiết bị.
- Công suất điện được tính theo công thức:

  $$
  P = \frac{A}{t} = UI
  $$

  hoặc khi biết điện trở:

  $$
  P = R I^2 = \frac{U^2}{R}
  $$

  Trong đó:
  - $ P $ là công suất điện (W).
  - $ U $ là hiệu điện thế (V).
  - $ I $ là cường độ dòng điện (A).
  - $ R $ là điện trở của thiết bị (Ω).

- **Đơn vị công suất điện**:
  - **Watt (W)**: Đơn vị chuẩn.
  - **Kilowatt (kW)**: 1 kW = 1000 W.

- **Công suất tiêu thụ của một số thiết bị điện phổ biến**:
  | Thiết bị       | Công suất (W) |
  |---------------|--------------|
  | Bóng đèn LED  | 5 - 20 W      |
  | Quạt điện     | 50 - 100 W    |
  | Tủ lạnh       | 100 - 400 W   |
  | Máy lạnh      | 1000 - 3000 W |
  | Bếp điện      | 1500 - 3000 W |

---

## **Tóm tắt**
- **Năng lượng điện** là công mà dòng điện thực hiện khi chạy qua một thiết bị.
- **Công thức tính năng lượng điện**: $ A = UIt = R I^2 t $.
- **Công suất điện** đo tốc độ tiêu thụ năng lượng của thiết bị.
- **Công thức tính công suất**: $ P = UI = R I^2 = \frac{U^2}{R} $.
- **Đơn vị**:
  - Năng lượng: J, Wh, kWh.
  - Công suất: W, kW.

---

🔹 **Ứng dụng**:  
- Tính toán tiền điện trong gia đình dựa trên công suất và thời gian sử dụng của thiết bị.
- Thiết kế hệ thống điện sao cho an toàn và hiệu quả.
- Lựa chọn thiết bị điện tiết kiệm năng lượng (công suất thấp nhưng hiệu suất cao).
