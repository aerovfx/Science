# **ĐIỆN TRỞ VÀ ĐỊNH LUẬT ÔM**

## **1. Điện trở**

- **Điện trở (R)** của một vật dẫn đặc trưng cho khả năng cản trở dòng điện của vật đó.
- Giá trị điện trở của một vật phụ thuộc vào:
  - **Bản chất của vật liệu**: Kim loại có điện trở nhỏ, chất bán dẫn có điện trở trung bình, chất cách điện có điện trở rất lớn.
  - **Kích thước hình học của dây dẫn**:
    - Điện trở tỷ lệ thuận với chiều dài $ l $ của dây dẫn.
    - Điện trở tỷ lệ nghịch với tiết diện ngang $ S $ của dây dẫn.

  Công thức xác định điện trở:

  $$
  R = \rho \frac{l}{S}
  $$

  Trong đó:
  - $ R $ là điện trở (Ω – Ohm).
  - $ \rho $ là điện trở suất của vật liệu (Ω·m).
  - $ l $ là chiều dài dây dẫn (m).
  - $ S $ là tiết diện ngang của dây dẫn (m²).

- **Đơn vị đo**: Ôm (Ω).

- **Dụng cụ đo**: **Ohm kế** hoặc **vôn kế – ampe kế** theo định luật Ôm.


---
## **3. Đường đặc trưng Vôn - Ampe**

- **Phát biểu**: ĐƯờng thẳng xuất phát từ gốc toạ độ và biểu diễn sự phụ thuộc giữa hiệu điện thế và cường độ dòng điện trong đoạn mạch thuần điện trở.
- **Vẽ hình: Mathplot.**




## **2. Định luật Ôm**

- **Phát biểu**: Cường độ dòng điện $ I $ chạy qua một vật dẫn tỷ lệ thuận với hiệu điện thế $ U $ đặt vào hai đầu vật dẫn và tỷ lệ nghịch với điện trở $ R $ của vật đó.

  Công thức định luật Ôm:

  $$
  I = \frac{U}{R}
  $$

  Trong đó:
  - $ I $ là cường độ dòng điện (A).
  - $ U $ là hiệu điện thế đặt vào hai đầu vật dẫn (V).
  - $ R $ là điện trở của vật dẫn (Ω).

- **Biểu diễn đồ thị**:
  - Nếu $ R $ không đổi, đồ thị biểu diễn sự phụ thuộc của $ I $ vào $ U $ là một đường thẳng đi qua gốc tọa độ.
  - Với các linh kiện như điện trở thuần, định luật Ôm luôn đúng.
  - Với một số linh kiện như diode, bóng đèn sợi đốt, điện trở có thể thay đổi theo dòng điện.

## **3. Nguyên nhân gây ra điện trở và ảnh hưởng của nhiệt độ lên điện trở**

### **Nguyên nhân gây ra điện trở**
- Điện trở của vật dẫn xuất phát từ **sự va chạm giữa các electron tự do với ion trong mạng tinh thể**.
- Khi các electron di chuyển, chúng bị cản trở bởi ion trong mạng tinh thể, gây ra **điện trở**.

### **Ảnh hưởng của nhiệt độ lên điện trở**
- Với **kim loại**: Khi nhiệt độ tăng, dao động nhiệt của ion trong mạng tinh thể tăng, làm tăng điện trở.
- Với **bán dẫn**: Khi nhiệt độ tăng, số lượng hạt tải điện tăng, làm điện trở giảm.
- Công thức mô tả sự thay đổi điện trở theo nhiệt độ:

  $$
  R = R_0 (1 + \alpha \Delta T)
  $$

  Trong đó:
  - $ R_0 $ là điện trở tại nhiệt độ ban đầu ($ T_0 $).
  - $ R $ là điện trở tại nhiệt độ $ T $.
  - $ \alpha $ là hệ số nhiệt điện trở của vật liệu.
  - $ \Delta T = T - T_0 $ là độ thay đổi nhiệt độ.

---

## **Tóm tắt**
- Điện trở đặc trưng cho khả năng cản trở dòng điện và phụ thuộc vào bản chất, kích thước vật liệu.
- Định luật Ôm: $ I = \frac{U}{R} $.
- Điện trở thay đổi theo nhiệt độ: Kim loại có điện trở tăng khi nhiệt độ tăng, bán dẫn thì ngược lại.

---

🔹 **Ứng dụng**: Định luật Ôm giúp thiết kế mạch điện, kiểm soát dòng điện trong thiết bị điện tử, và xác định điện trở phù hợp trong các linh kiện điện tử.
