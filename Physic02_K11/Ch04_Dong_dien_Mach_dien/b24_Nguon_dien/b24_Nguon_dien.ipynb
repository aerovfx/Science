# **NGUỒN ĐIỆN**

## **1. Nguồn điện. Suất điện động của nguồn điện**

### **a) Khái niệm nguồn điện**
- **Nguồn điện** là thiết bị có khả năng **duy trì một hiệu điện thế** giữa hai cực của nó bằng cách thực hiện công lên các hạt mang điện.
- Nguồn điện có thể là **pin, ắc quy, máy phát điện, pin mặt trời, v.v.**
- Trong một nguồn điện, năng lượng từ các dạng khác nhau (hóa năng, cơ năng, quang năng, v.v.) được chuyển hóa thành điện năng.

### **b) Suất điện động của nguồn điện**
- **Suất điện động (sđđ) của nguồn điện**, ký hiệu là **$ \mathcal{E} $**, đặc trưng cho khả năng thực hiện công của nguồn điện để duy trì dòng điện trong mạch.
- Được xác định bằng công thức:

  $$
  \mathcal{E} = \frac{A}{q}
  $$

  Trong đó:
  - $ \mathcal{E} $ là suất điện động (V).
  - $ A $ là công của lực lạ thực hiện để dịch chuyển điện tích (J).
  - $ q $ là điện tích dịch chuyển (C).

- **Đơn vị**: Vôn (V).

- **Suất điện động khác với hiệu điện thế**:  
  - Hiệu điện thế giữa hai cực của nguồn có thể thay đổi tùy theo dòng điện chạy qua.
  - Suất điện động là một đại lượng đặc trưng cố định cho từng nguồn điện.

---

## **2. Ảnh hưởng của điện trở trong của nguồn điện lên hiệu điện thế giữa hai cực của nguồn**

### **a) Điện trở trong của nguồn điện**
- Trong nguồn điện luôn tồn tại một phần **điện trở trong**, ký hiệu là $ r $, do vật liệu và cấu trúc của nguồn gây ra.
- Khi nguồn điện hoạt động, điện trở trong sẽ làm giảm hiệu điện thế giữa hai cực của nguồn.

### **b) Công thức liên hệ giữa suất điện động, hiệu điện thế và điện trở trong**
- Hiệu điện thế giữa hai cực của nguồn khi có dòng điện chạy qua:

  $$
  U = \mathcal{E} - Ir
  $$

  Trong đó:
  - $ U $ là hiệu điện thế giữa hai cực của nguồn (V).
  - $ \mathcal{E} $ là suất điện động của nguồn (V).
  - $ I $ là cường độ dòng điện qua nguồn (A).
  - $ r $ là điện trở trong của nguồn điện (Ω).

- Khi mạch **không tải** ($ I = 0 $), ta có:

  $$
  U_0 = \mathcal{E}
  $$

  Nghĩa là hiệu điện thế giữa hai cực của nguồn bằng suất điện động khi không có dòng điện.

### **c) Ảnh hưởng của điện trở trong lên hiệu điện thế**
- Nếu **$ r $ lớn**, hiệu điện thế thực tế của nguồn điện **giảm nhiều** khi cấp dòng điện.
- Nếu **$ r $ nhỏ**, hiệu điện thế thực tế **gần bằng suất điện động**.
- Khi dòng điện **tăng quá lớn**, hiệu điện thế có thể giảm mạnh, gây mất ổn định cho thiết bị sử dụng điện.

---

## **Tóm tắt**
- **Nguồn điện** duy trì dòng điện bằng cách sinh công lên các hạt mang điện.
- **Suất điện động ($ \mathcal{E} $)** đo khả năng thực hiện công của nguồn điện.
- **Điện trở trong ($ r $)** làm giảm hiệu điện thế giữa hai cực của nguồn khi có dòng điện chạy qua.
- **Công thức liên hệ**: $ U = \mathcal{E} - Ir $.

---

🔹 **Ứng dụng**:  
- Khi thiết kế mạch điện, cần chọn nguồn điện có **suất điện động đủ lớn** và **điện trở trong nhỏ** để đảm bảo điện áp cung cấp ổn định.
- Trong ắc quy và pin, điện trở trong tăng theo thời gian, làm giảm hiệu suất hoạt động.
