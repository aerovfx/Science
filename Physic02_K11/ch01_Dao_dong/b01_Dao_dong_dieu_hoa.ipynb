### I. NHỮNG ĐẶC ĐIỂM CỦA DAO ĐỘNG CƠ

1. **Thí nghiệm**

**Chuẩn bị:**  
Sử dụng **con lắc lò xo** hoặc **con lắc đơn** (Hình 1.1).

**Tiến hành:**  
- Treo một vật nhỏ, nặng vào đầu tự do của một lò xo nhẹ (Hình 1.1a) hoặc một dây nhẹ không dãn ta có con lắc lò xo hoặc con lắc đơn (Hình 1.1b).  
- Thực hiện:  
  1. Xác định vị trí cân bằng của vật.  
  2. Kéo vật lệch khỏi vị trí cân bằng rồi thả ra cho nó chuyển động.  
  3. Quan sát chuyển động của mỗi vật và cho nhận xét về đặc điểm chung của chúng.


Nêu những ví d ụ vê dao động c ơ mà em biết.

2. **Dao động cơ**

- Vật chuyển động qua lại quanh vị trí cân bằng. Chuyển động như vậy gọi là **dao động cơ**.  
- Dao động cơ của một vật có thể là **tuần hoàn** hoặc **không tuần hoàn**.  

**Định nghĩa:**  
- Nếu sau những khoảng thời gian bằng nhau, vật trở lại vị trí cũ theo hướng cũ thì dao động của vật đó là **tuần hoàn**.  

**Ví dụ:**  
- Dao động của **con lắc đồng hồ** là tuần hoàn.  
- Dao động của **cành cây đu đưa khi gió thổi** là không tuần hoàn.  

**Ghi chú:**  
- Dao động tuần hoàn có thể có mức độ phức tạp khác nhau.  
- Dao động tuần hoàn đơn giản nhất là **dao động điều hoà**.


### II. DAO ĐỘNG ĐIỀU HÒA

1. **Đồ thị của dao động điều hoà**

Đường cong trên **Hình 1.2** là *đồ thị dao động* của con lắc: nó cho biết vị trí \(x\) của quả cầu trên
trục \(x\) tại những thời điểm khác nhau \(t\). Đường cong này có dạng **hình sin**.

2. **Phương trình của dao động điều hoà**

Tương ứng với đồ thị hình sin là **hàm sin/cosin**. Ta chọn mô tả bằng cosin:
\[
x(t) = A\cos\!\big(\omega t + \varphi\big) \tag{1.1}
\]
Phương trình (1.1) gọi là **phương trình dao động điều hoà**.

**Ý nghĩa các hằng số:**
- \(A>0\): **biên độ** (độ lệch cực đại, đơn vị cùng với \(x\))
- \(\omega>0\): **tần số góc** \([{\rm rad/s}]\), liên hệ \(T=\dfrac{2\pi}{\omega}\), \(f=\dfrac{1}{T}=\dfrac{\omega}{2\pi}\)
- \(\varphi\): **pha ban đầu** \([{\rm rad}]\) tại \(t=0\)
- **Pha tức thời**: \(\theta(t)=\omega t+\varphi\)
- **Li độ** (vị trí tức thời): \(x(t)=A\cos\theta(t)\)

---

#### Bài tập mẫu
Cho phương trình:
\[
x(t)=A\cos(\omega t+\varphi)\quad(\text{đơn vị: cm})
\]
Hãy xác định:
1) **Biên độ** và **pha ban đầu**.  
2) **Pha** và **li độ** tại \(t=2\,\text{s}\).

**Lời giải tổng quát (áp dụng cho mọi số liệu):**
- a) Từ phương trình, đọc ngay:
  - \(A\) là hệ số trước \(\cos\)  \(\Rightarrow\) **biên độ** \(=A\).
  - \(\varphi\) là số hạng cộng với \(\omega t\)  \(\Rightarrow\) **pha ban đầu** \(=\varphi\).
- b) Tại \(t=2\,\text{s}\):
  - **Pha**: \(\theta(2)=\omega\cdot 2+\varphi=2\omega+\varphi\) (có thể quy về \([0,2\pi)\) bằng cách cộng/trừ bội \(2\pi\)).
  - **Li độ**: 
    \[
    x(2)=A\cos\!\big(2\omega+\varphi\big).
    \]

> *Ghi chú:* Nếu đề cho cụ thể, ví dụ \(x=2\cos(4\pi t+\varphi)\) (cm) thì:
> - \(A=2\) cm; \(\omega=4\pi\) rad/s \(\Rightarrow T=\dfrac{2\pi}{4\pi}=0{,}5\) s; \(f=2\) Hz.
> - Với pha ban đầu cho bởi \(\varphi\) (ví dụ \(\varphi=-\dfrac{\pi}{2}\) nếu đề nêu như vậy), khi \(t=2\) s:
>   \[
>   \theta(2)=4\pi\cdot 2+\varphi=8\pi+\varphi\equiv \varphi \pmod{2\pi},\quad
>   x(2)=2\cos(\varphi).
>   \]

**Kết luận:** Mọi bài dạng này đều quy về việc **đọc \(A,\omega,\varphi\)** từ phương trình, sau đó thế vào
\(\theta(t)=\omega t+\varphi\) và \(x(t)=A\cos\theta(t)\). 


### Hình 1.3 — Phân tích sâu (dao động điều hoà của con lắc đơn)

#### 1) Mô tả dao động điều hoà của con lắc đơn (theo đồ thị)
- Đồ thị cho thấy quả cầu bắt đầu ở **vị trí cực đại dương** tại \(t=0\) (li độ dương lớn nhất), sau đó chuyển động về phía **vị trí cân bằng** \(x=0\), đi qua đó xuống phía âm, đạt **cực tiểu** tại một thời điểm sau đó rồi quay trở lại.  
- Hình dạng đường cong là **hình sin/cosin** điển hình của dao động điều hoà.  
- Đặc điểm tổng quát của chuyển động:  
  - Dao động **tuần hoàn**: chuyển động lặp lại sau một khoảng thời gian \(T\) (chu kỳ).  
  - Quả cầu dao động quanh **vị trí cân bằng** \(x=0\).  
  - Tại các **cực trị** (biên), vận tốc bằng 0, thế năng cực đại; khi qua vị trí cân bằng, tốc độ cực đại, động năng cực đại.  
- Từ đồ thị suy ra: đường bắt đầu ở cực đại dương → phương trình có dạng cosin với pha ban đầu \(\varphi\) bằng \(0\) (hoặc một bội \(2\pi\)).

---

#### 2) Xác định biên độ và li độ tại \(t=0,\;0{,}5\ \text{s},\;2{,}0\ \text{s}\)

**Bước 1 — Nhận dạng phương trình**  
Từ đồ thị:  
- Biên độ (độ lệch lớn nhất trên trục \(x\)) là \(A \approx 40\ \text{cm}\).  
- Đồ thị có giá trị cực đại dương tại \(t=0\) ⇒ chọn dạng cosin với pha ban đầu \(\varphi=0\).  
Do đó ta lấy phương trình mô tả:
\[
\boxed{\,x(t)=A\cos(\omega t+\varphi)=40\cos(\omega t)\ \text{(cm)}\,}
\]

**Bước 2 — Xác định chu kỳ \(T\) và tần số góc \(\omega\)**  
- Quan sát: đồ thị đạt **cực tiểu** tại \(t\approx 2{,}0\ \text{s}\). Vì với cosin, cực đại tại \(t=0\) và cực tiểu xuất hiện sau nửa chu kỳ \(T/2\), suy ra
  \[
  T/2 \approx 2{,}0\ \Rightarrow\ T \approx 4{,}0\ \text{s}.
  \]
- Tần số: \(f=\dfrac{1}{T}=\dfrac{1}{4}=0{,}25\ \text{Hz}.\)  
- Tần số góc:
  \[
  \omega = \dfrac{2\pi}{T} = \dfrac{2\pi}{4} = \dfrac{\pi}{2}\ \text{rad/s}.
  \]

Vậy phương trình đầy đủ ta dùng cho tính toán là
\[
\boxed{\,x(t)=40\cos\!\Big(\tfrac{\pi}{2}\,t\Big)\ \text{(cm)}\,}
\]

**Bước 3 — Tính các giá trị yêu cầu**

- \(t=0\):  
  \[
  x(0)=40\cos(0)=40\ \text{cm}.
  \]
  (Đúng với đồ thị: bắt đầu ở cực đại dương.)

- \(t=0{,}5\ \text{s}\):  
  \[
  x(0{,}5)=40\cos\!\big(\tfrac{\pi}{2}\cdot 0{,}5\big)=40\cos\!\big(\tfrac{\pi}{4}\big)
         =40\cdot\dfrac{\sqrt{2}}{2}\approx 40\cdot 0{,}7071\approx 28{,}3\ \text{cm}.
  \]

- \(t=2{,}0\ \text{s}\):  
  \[
  x(2)=40\cos\!\big(\tfrac{\pi}{2}\cdot 2\big)=40\cos(\pi)=40\cdot(-1)=-40\ \text{cm}.
  \]

**Tóm tắt kết quả:**  
- **Biên độ:** \(A=40\ \text{cm}\).  
- **Phương trình (gợi ý):** \(x(t)=40\cos\!\big(\tfrac{\pi}{2}t\big)\) cm.  
- **Giá trị li độ:**  
  - \(x(0)=40\ \text{cm}\).  
  - \(x(0{,}5)\approx 28{,}3\ \text{cm}\).  
  - \(x(2{,}0)=-40\ \text{cm}\).

---

#### Ghi chú phân tích sâu (độ chính xác và sai số quan sát)
- Kết luận trên dựa vào việc đọc đồ thị: cực đại ban đầu ≈ \(+40\) cm; cực tiểu ≈ \(-40\) cm tại \(t\approx 2\) s ⇒ suy ra \(T\approx4\) s. Nếu đồ thị thực tế có lệch nhỏ (vì vẽ tay hoặc do ảnh chụp), các con số làm tròn có thể khác vài phần trăm.  
- Nếu bạn muốn, mình có thể:  
  - vẽ lại **đồ thị lý thuyết** của \(x(t)=40\cos\big(\tfrac{\pi}{2}t\big)\) để so sánh trực tiếp với ảnh; hoặc  
  - điều chỉnh pha \(\varphi\) hoặc \(A\) để khớp chính xác hơn với đường màu trong ảnh (nếu bạn muốn mình đo tọa độ điểm ảnh chính xác từ file ảnh).
