```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Hằng số
k = 8.99e9  # Hằng số Coulomb (N·m²/C²)

# Điện tích điểm: Tính toán điện trường và điện thế theo khoảng cách r
def electric_field(Q, r):
    return k * Q / r**2

# Tham số hóa dữ liệu
r_values = np.linspace(0.1, 5, 100)  # Khoảng cách từ 0.1m đến 5m

# Điện tích ban đầu
Q = 1e-6  # 1 microCoulomb

# Bài tập 1: Điện trường của điện tích điểm
E_values = electric_field(Q, r_values)

plt.figure(figsize=(8, 5))
ax1 = plt.subplot()
plt.plot(r_values, E_values, label='E(r) của điện tích điểm', color='b')
plt.xlabel('Khoảng cách r (m)')
plt.ylabel('Điện trường E (N/C)')
plt.title('Điện trường của điện tích điểm')
plt.legend()

# Tạo thanh trượt
axQ = plt.axes([0.25, 0.1, 0.65, 0.03])
slider_Q = Slider(axQ, 'Điện tích Q', -2e-3, 1e-3, valinit=Q)

# Hàm cập nhật khi giá trị điện tích thay đổi
def update(val):
    Q = slider_Q.val
    E_values = electric_field(Q, r_values)
    ax1.clear()
    plt.plot(r_values, E_values, label='E(r) của điện tích điểm', color='b')
    plt.xlabel('Khoảng cách r (m)')
    plt.ylabel('Điện trường E (N/C)')
    plt.title('Điện trường của điện tích điểm')
    plt.legend()
    plt.draw_idle()

# Cài đặt sự kiện cho thanh trượt
slider_Q.on_changed(update)

plt.show()
```
