```python
import numpy as np
import matplotlib.pyplot as plt

# Hằng số Coulomb
k = 8.99e9 # N·m²/C²

# Điện tích điểm: Tính toán lực điện theo khoảng cách r
def electric_force(Q, r):
    return k * Q**2 / r**2

# Tham số hóa dữ liệu
r_values = np.linspace(0.1, 5, 100) # Khoảng cách từ 0.1m đến 5m
Q = 1e-6 # 1 microCoulomb (đơn vị điện tích)

# Tính toán lực điện tại mỗi khoảng cách r
F_values = electric_force(Q, r_values)

# Vẽ đồ thị lực điện theo khoảng cách r
plt.figure(figsize=(8, 5))
ax1 = plt.subplot()
plt.plot(r_values*1000, F_values*10000, label='Lực điện (N)', color='b')
plt.xlabel('Khoảng cách r (mm)')
plt.ylabel('Lực F (kN)')
plt.title('Lực điện của điện tích điểm theo khoảng cách')
plt.legend()

# Thêm lưới grid
ax1.grid(True)

# Hiện thị đồ thị
plt.show()
```
