```python
import numpy as np
import matplotlib.pyplot as plt

# Hằng số Coulomb
k = 8.99e9  # N·m²/C²

# Điện tích điểm: Tính toán điện trường theo khoảng cách r
def electric_field(Q, r):
    return k * Q / r**2

# Tham số hóa dữ liệu
r_values = np.linspace(0.1, 5, 100)  # Khoảng cách từ 0.1m đến 5m
Q = 1e-6  # 1 microCoulomb (đơn vị điện tích)

# Tính toán điện trường tại mỗi khoảng cách r
E_values = electric_field(Q, r_values)

# Vẽ đồ thị điện trường theo khoảng cách r
plt.figure(figsize=(8, 5))
ax1 = plt.subplot()
plt.plot(r_values*1000, E_values*10000, label='E(r) của điện tích điểm', color='b')
plt.xlabel('Khoảng cách r (mm)')
plt.ylabel('Điện trường E (kN/C)')
plt.title('Điện trường của điện tích điểm')
plt.legend()

# Vẽ đồ thị điện thế theo khoảng cách r
r_values_potential = np.linspace(0.1, 5, 100)
V_values = k * Q / r_values_potential
ax2 = plt.subplot()
plt.plot(r_values_potential*1000, V_values*10000, label='V(r) của điện tích điểm', color='g')
plt.xlabel('Khoảng cách r (mm)')
plt.ylabel('Điện thế V (kV)')
plt.title('Điện thế của điện tích điểm')
plt.legend()

# Hiện thị cả hai đồ thị cùng lúc
plt.tight_layout()
plt.show()
```
