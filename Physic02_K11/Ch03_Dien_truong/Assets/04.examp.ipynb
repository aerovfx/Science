### Bài tập: Tương tác điện giữa electron và proton của nguyên tử Hidrogen
* Nguyên tử Hidrogen là một trong những nguyên tử đơn giản nhất, chỉ có 1 electron và 1 proton. Electron và proton được giữ lại bởi lực điện tĩnh, còn gọi là lực Coulomb.
* Mời bạn thực hiện bài tập sau:
* Bước 1: Tính toán lực Coulomb
* Lực Coulomb giữa 2 điện tích điểm (electron và proton) được tính bằng công thức:
*   F = k * (q1 * q2) / r^2
* trong đó:
F là lực Coulomb
k là hằng số Coulomb (6,25 x 10^-11 N m^2 C^-2)
q1 và q2 là điện tích của electron và proton (both là -1,6 x 10^-19 C)
r là khoảng cách giữa electron và proton
* Bước 2: Vẽ đồ thị lực Coulomb theo khoảng cách
Vì lực Coulomb giảm nhanh khi khoảng cách tăng, chúng ta có thể vẽ đồ thị lực Coulomb theo khoảng cách.

```python
import numpy as np
import matplotlib.pyplot as plt

# Hằng số Coulomb
k = 6.25e-11 # N m^2 C^-2

# Điện tích của electron và proton
q1 = q2 = -1.6e-19 # C

# Khoảng cách từ 0 đến 10 Å (Angstrom)
r = np.linspace(0, 10, 100) # Å

# Tính toán lực Coulomb
F = k * (q1 * q2) / r**2    # r khác 0  

# Vẽ đồ thị
plt.plot(r, F)
plt.xlabel('Khoảng cách (Å)')
plt.ylabel('Lực Coulomb (N)')
plt.title('Tương tác điện giữa electron và proton của nguyên tử Hidrogen')
plt.grid(True)
plt.show()
```
