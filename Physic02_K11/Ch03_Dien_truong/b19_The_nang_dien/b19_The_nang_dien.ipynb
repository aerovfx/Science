# **Thế Năng Điện**

## **1. Công của lực điện**
- **Công của lực điện** là công mà lực điện thực hiện khi di chuyển một điện tích trong một điện trường. Công này phụ thuộc vào điện tích, cường độ điện trường và quãng đường di chuyển của điện tích.
- Để tính công của lực điện, ta sử dụng công thức:

  $$
  A = \mathbf{F} \cdot \mathbf{d} = q \mathbf{E} \cdot \mathbf{d}
  $$

  Trong đó:
  - $A$ là công của lực điện (J).
  - $q$ là điện tích thử (C).
  - $\mathbf{E}$ là cường độ điện trường (V/m).
  - $\mathbf{d}$ là vectơ dịch chuyển của điện tích (m).
  
- Khi điện tích di chuyển trong điện trường, công của lực điện sẽ là dương nếu điện tích di chuyển theo phương của lực, và là âm nếu di chuyển ngược lại.

- Công của lực điện có thể được tính thông qua sự thay đổi trong thế năng điện của hệ thống.

## **2. Thế năng của một điện tích trong điện trường**
- **Thế năng điện** của một điện tích trong điện trường là năng lượng tiềm tàng mà điện tích đó có được do vị trí của nó trong điện trường.
- Thế năng điện của điện tích $q$ tại một điểm trong điện trường có thể tính bằng công thức:

  $$
  W = qV
  $$

  Trong đó:
  - $W$ là thế năng điện của điện tích (J).
  - $q$ là độ lớn của điện tích thử (C).
  - $V$ là hiệu điện thế tại điểm điện tích đang đặt (V).

- **Hiệu điện thế $V$** tại một điểm trong điện trường là công của lực điện thực hiện để đưa một đơn vị điện tích từ vô cùng về điểm đó. Nó được tính bằng:

  $$
  V = \frac{W}{q}
  $$

  Trong đó:
  - $W$ là công thực hiện để mang điện tích $q$ từ vô cùng đến điểm có hiệu điện thế $V$.
  - $q$ là độ lớn của điện tích thử.

- Nếu điện tích di chuyển trong điện trường, thế năng của nó thay đổi. Công của lực điện khi điện tích di chuyển từ một điểm có thế năng $W_1$ đến một điểm có thế năng $W_2$ là:

  $$
  A = W_1 - W_2
  $$

  Lực điện thực hiện công này là lực bảo toàn, tức là tổng năng của hệ bao gồm thế năng điện và động năng của điện tích luôn được bảo toàn trong một hệ kín.

---

### Ứng dụng:
- **Pin và acquy**: Các thiết bị này dựa vào sự thay đổi thế năng điện trong quá trình chuyển hóa giữa các dạng năng lượng khác nhau.
- **Điện trường trong máy gia tốc**: Công của lực điện giúp tăng tốc hạt trong các gia tốc tuyến tính và máy gia tốc hạt.
- **Thiết bị lưu trữ năng lượng**: Các tụ điện là ví dụ điển hình trong việc lưu trữ năng lượng điện dưới dạng thế năng.
