# **KHÁI NIỆM ĐIỆN TRƯỜNG**

## **1. Khái niệm điện trường**
- **Điện trường** là một dạng vật chất tồn tại xung quanh các điện tích và tác dụng lực điện lên các điện tích khác đặt trong nó.
- Điện trường là môi trường truyền tương tác điện giữa các điện tích.
- Nếu một điện tích thử \( q \) đặt tại một điểm mà xuất hiện lực điện tác dụng lên nó, thì tại điểm đó có điện trường.

---

## **2. Cường độ điện trường**
- **Cường độ điện trường** tại một điểm đặc trưng cho tác dụng lực của điện trường tại điểm đó.
- Cường độ điện trường là đại lượng vector, được xác định bởi công thức:

$$
E = \frac{F}{q}
$$

Trong đó:
- \( E \) là cường độ điện trường (V/m).
- \( F \) là lực điện tác dụng lên điện tích thử \( q \) (N).
- \( q \) là điện tích thử (C).

- **Đơn vị**: Vôn trên mét (V/m).
- **Phương và chiều của vector cường độ điện trường**:
  - **Do điện tích dương** tạo ra: $\vec{E}$ có phương **hướng ra xa** điện tích.
  - **Do điện tích âm** tạo ra: $\vec{E}$ có phương **hướng vào** điện tích.

- **Cường độ điện trường của điện tích điểm** được tính bằng:

$$
E = \frac{k |Q|}{r^2}
$$

Trong đó:
- $ k = 9 \times 10^9 $ (Nm²/C²) là hằng số điện.
- $ Q $ là điện tích tạo ra điện trường (C).
- $ r $ là khoảng cách từ điện tích đến điểm xét (m).

---

## **3. Điện phổ**
- **Điện phổ** là hình ảnh trực quan của điện trường, biểu diễn bằng các **đường sức điện**.
- **Đường sức điện** là đường cong vẽ trong điện trường sao cho tiếp tuyến tại mỗi điểm có phương trùng với phương của vector cường độ điện trường tại điểm đó.
- **Tính chất của đường sức điện**:
  1. **Qua mỗi điểm trong điện trường chỉ có một đường sức**.
  2. **Các đường sức của điện trường tĩnh không cắt nhau**.
  3. **Chiều của đường sức điện**:  
     - Xuất phát từ **điện tích dương** và kết thúc ở **điện tích âm**.
  4. **Độ mau thưa của đường sức thể hiện độ mạnh yếu của điện trường**:  
     - **Dày**: điện trường mạnh.
     - **Thưa**: điện trường yếu.
  5. **Điện trường đều**: Các đường sức **song song, cùng chiều, cách đều nhau** (ví dụ: giữa hai bản tụ điện phẳng song song).

---

🔹 **Ứng dụng**: Điện trường đóng vai trò quan trọng trong các thiết bị điện tử, cảm biến, tụ điện, và các nghiên cứu về sóng điện từ.
