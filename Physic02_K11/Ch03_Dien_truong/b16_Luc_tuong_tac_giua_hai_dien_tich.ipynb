# **LỰC TƯƠNG TÁC ĐIỆN – ĐỊNH LUẬT COULOMB**

## **📌 Ứng dụng thực tế**
- Lực tương tác giữa các điện tích đóng vai trò quan trọng trong **công nghệ điện tử, vật liệu, y sinh, và viễn thông**.
- **Ví dụ ứng dụng**:
  - **Máy photocopy, máy in laser**: Hoạt động dựa trên nguyên lý lực hút giữa các điện tích trái dấu.
  - **Lọc bụi tĩnh điện**: Dùng để loại bỏ hạt bụi trong không khí bằng cách tích điện cho chúng.
  - **Tụ điện**: Dựa vào lực tương tác giữa các điện tích để lưu trữ năng lượng trong mạch điện.

---

## **1️⃣ Lực hút và lực đẩy giữa các điện tích**
- **Điện tích** là thuộc tính của vật chất tạo ra **tương tác điện** giữa các hạt mang điện.
- Có **hai loại điện tích**:
  - **Điện tích dương** (\(+\)) và **điện tích âm** (\(-\)).
- Quy tắc cơ bản của lực tương tác điện:
  - **Các điện tích cùng dấu** → **đẩy nhau**.
  - **Các điện tích khác dấu** → **hút nhau**.
- **Lực điện** là một **đại lượng vector**, có phương trùng với đường nối hai điện tích.

---

## **2️⃣ Định luật Coulomb**
- **Phát biểu**: Lực hút hay lực đẩy giữa hai điện tích điểm đặt trong chân không có:
  - **Phương**: Trùng với đường nối hai điện tích.
  - **Chiều**: Phụ thuộc vào dấu của hai điện tích.
  - **Độ lớn**: **Tỉ lệ thuận** với tích độ lớn hai điện tích và **tỉ lệ nghịch** với bình phương khoảng cách giữa chúng.

### **🔹 Công thức định luật Coulomb**

$$
F = k \frac{|Q_1 Q_2|}{r^2}
$$

Trong đó:
- $ F $ – Lực tương tác điện (N).
- $ k = 9 \times 10^9 $ (Nm²/C²) – Hằng số Coulomb.
- $ Q_1, Q_2 $ – Độ lớn của hai điện tích (C).
- $ r \$ – Khoảng cách giữa hai điện tích (m).

### **🔹 Hằng số điện môi**
- Nếu đặt điện tích trong **môi trường cách điện** thay vì chân không, lực tương tác giảm đi $\varepsilon $ lần:

$
F = \frac{k |Q_1 Q_2|}{\varepsilon r^2}
$

- $ \varepsilon $ là **hằng số điện môi**, đặc trưng cho khả năng cách điện của môi trường.

---

## **3️⃣ Bài tập về định luật Coulomb**
### **📝 Dạng 1: Tính lực tương tác giữa hai điện tích**
**Ví dụ 1:**  
Hai điện tích điểm \( Q_1 = 5 \times 10^{-6} \) C và \( Q_2 = -3 \times 10^{-6} \) C đặt cách nhau \( 10 \) cm trong chân không. Tính lực tương tác giữa chúng.

**Giải:**  
Sử dụng công thức:

$$
F = k \frac{|Q_1 Q_2|}{r^2}
$$

Thay số:
$$
F = 9 \times 10^9 \times \frac{(5 \times 10^{-6}) (3 \times 10^{-6})}{(0.1)^2}
$$

$$
F = 13.5 \text{ N} \quad (\text{lực hút do hai điện tích trái dấu})
$$

---

### **📝 Dạng 2: Xác định khoảng cách giữa hai điện tích khi biết lực tương tác**
**Ví dụ 2:**  
Hai điện tích điểm $ Q_1 = 4 \times 10^{-6} $ C và $ Q_2 = 6 \times 10^{-6} $ C hút nhau với lực $ 2 $ N trong chân không. Tìm khoảng cách giữa chúng.

**Giải:**  
Dùng công thức:

$$
r = \sqrt{\frac{k |Q_1 Q_2|}{F}}
$$

Thay số:
$$

r = \sqrt{\frac{(9 \times 10^9) \times (4 \times 10^{-6}) \times (6 \times 10^{-6})}{2}}

$$

$$
r \approx 0.18 \text{ m}
$$

---

## **📌 Tổng kết**
- **Lực điện** có thể là **lực hút hoặc lực đẩy** tùy vào dấu của điện tích.
- **Định luật Coulomb** mô tả lực tương tác giữa hai điện tích điểm.
- **Hằng số điện môi** ảnh hưởng đến lực tương tác trong các môi trường khác nhau.
- **Bài tập** về Coulomb giúp tính toán lực, điện tích, hoặc khoảng cách giữa các điện tích.
