# **Điện Thế**

## **1. Điện thế tại 1 điểm trong điện trường**

- **Điện thế** tại một điểm trong điện trường là công mà lực điện thực hiện để đưa một đơn vị điện tích từ vô cùng đến điểm đó mà không làm thay đổi chuyển động của điện tích. Nói cách khác, điện thế là một đại lượng vô hướng đặc trưng cho năng lượng tiềm tàng của điện tích trong điện trường.
  
- Điện thế tại một điểm trong điện trường có thể tính bằng công thức:

  $$
  V = \frac{A}{q}
  $$

  Trong đó:
  - $V$ là điện thế tại điểm đó (V).
  - $A$ là công của lực điện thực hiện khi đưa điện tích $q$ từ vô cùng đến điểm đó (J).
  - $q$ là điện tích thử (C).

- **Đơn vị** của điện thế là Vôn (V), với 1 V = 1 J/C. Tức là, điện thế 1 V tương đương với 1 Joule năng lượng được cung cấp cho mỗi Coulomb điện tích.

- Điện thế là đại lượng vô hướng, nhưng nó có thể thay đổi tùy theo vị trí của điểm trong điện trường. Trong các hệ thống điện trường, mỗi điểm sẽ có một giá trị điện thế khác nhau.

## **2. Mối liên hệ giữa điện thế và cường độ điện trường**

- Cường độ điện trường và điện thế có mối liên hệ mật thiết với nhau. Mối quan hệ này thể hiện qua sự thay đổi của điện thế dọc theo phương của cường độ điện trường.

- Mối liên hệ giữa cường độ điện trường $\mathbf{E}$ và điện thế $V$ được biểu diễn bằng công thức:

  $$
  \mathbf{E} = - \nabla V
  $$

  Trong đó:
  - $\mathbf{E}$ là cường độ điện trường (V/m).
  - $\nabla V$ là gradient của điện thế, tức là sự thay đổi của điện thế theo không gian. Gradient thể hiện sự thay đổi nhanh chậm của điện thế theo từng hướng.

- Trong trường hợp điện trường đều, cường độ điện trường $\mathbf{E}$ và sự thay đổi của điện thế giữa hai điểm có thể được tính đơn giản bằng công thức:

  $$
  E = \frac{V}{d}
  $$

  Trong đó:
  - $E$ là cường độ điện trường (V/m).
  - $V$ là hiệu điện thế giữa hai điểm.
  - $d$ là khoảng cách giữa hai điểm đó (m).

- Mối liên hệ này cho thấy điện trường càng mạnh (cường độ điện trường càng lớn) thì điện thế thay đổi nhanh chóng giữa các điểm trong điện trường. Nếu điện trường đều, điện thế thay đổi đều theo khoảng cách giữa hai điểm.
